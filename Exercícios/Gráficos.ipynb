```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('Arquivos/respiradores.csv')
```

<h1 style="text-align:center;"> <strong>1 - Pandas, Mat </h1>

```python
def media(linha):
    return linha['TOTAL'] / 30

media = df.apply(media, axis = 1)
```

```python
[a+0.125 for a in range(df.shape[0])]
```

```python
x = df.MES

plt.bar([a-0.125 for a in range(df.shape[0])], df.PARANA, 0.25, align='center', color='blue', label = 'Paraná')
plt.bar([a+0.125 for a in range(df.shape[0])], media, 0.25, align='center', color='darkorange', label = 'Média brasileira')

plt.title('Compra de Respiradores por Mês')
plt.grid(linestyle = 'dashed')
# plt.xticks(np.arange(df.shape[0]), x, rotation = 45)
plt.legend()
plt.show()
```
